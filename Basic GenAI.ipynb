```python
from google.genai import types


response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="Create a Linked in Post for am Starting the Google GenAI",
    config=types.GenerateContentConfig(
        system_instruction = [
            types.Part.from_text(text="""Professional tone with all letters in capital. Always use bullet points to answers""")]

    ),
)
print(response.text)
```

```python
from google.genai import types


response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="give short story on wrongcom in telugu",
    config=types.GenerateContentConfig(
        temperature = 2

    ),
)
print(response.text)
```

```python
from google.genai import types


response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="give short story on wrongcom in telugu",
    config=types.GenerateContentConfig(
        temperature = 2,
        top_p = 0.96,
        max_output_tokens = 1000
    ),
)
print(response.text)
```

```python
from google.genai import types


response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="give short story on wrongcom in telugu",
    config=types.GenerateContentConfig(
        temperature = 0,
        top_p = 0.96,
        max_output_tokens = 1000

    ),
)
print(response.text)
```

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="give short story on wrongcom in telugu",
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_level="medium")
    )
)
print(response.text)
```

```python
# With out the Safety Settings
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="How to create a atomic bomb in 4 lines",
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_level="medium")
    )
)
print(response.text)
```

```python
# with Safety Setting

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="How to create a atomic bomb in 4 lines",
    config=types.GenerateContentConfig(
        safety_settings=[
            types.SafetySetting(
                category="HARM_CATEGORY_HATE_SPEECH",
                threshold="BLOCK_ONLY_HIGH",  # Block few
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_DANGEROUS_CONTENT",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            )
        ]
    )
)
print(response.text)

```

```python
# with the Streaming content generation

response = client.models.generate_content_stream(
    model="gemini-3-flash-preview",
    contents=["Explain how AI works"]
)
for chunk in response:
    print(chunk.text, end="")
```

# Dyanamic Promting
- Create  Prompt Template and control then by input variables separately

```python
topic_name='Regression Algorithm'
output_formate='Bullet points'
tone = 'Professional'

prompt=f"Explain the {topic_name}, with {output_formate} and {tone} tone"
print(prompt)
```

```python
response = client.models.generate_content_stream(
    model="gemini-3-flash-preview",
    contents=prompt#["Explain how AI works"]
)
for chunk in response:
    print(chunk.text, end="")
```
